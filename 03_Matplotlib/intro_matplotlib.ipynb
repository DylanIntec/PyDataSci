Matplotlib is one of the most widly used graphing libraries in Python. It can be confusing to start. 

```python
import matplotlib.pyplot as plt
import numpy as np
```

Creation of a line graph:

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])
```

The image below shows the basic anatomy of a matplotlib graph. The most important objects are fig and ax. 

<img src="../_images/fig_map.png">

To modify the axes you use set commands:

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])
ax.set_xlabel('Time [years]')
ax.set_ylabel('Digital growth [%]')
ax.set_title('Growth in Digital')
```

Anything on the graph can be set to any value. Here are more common objects to set:

<img src="../_images/fig_map2.png">

```python
fig, ax = plt.subplots()
ax.scatter([1, 2, 3, 4], [1, 4, 2, 3], marker='o', c='w', edgecolor='b', label='Perth')
ax.scatter([1, 2, 3, 4], [1.4, 2, 2.5, 3.8], marker='*', c='w', edgecolor='r', label='GID')
ax.set_xlim((0, 5))
ax.set_ylim((0, 5))
ax.set_xticks(np.arange(0.5, 5, 1), minor=True)
ax.set_yticks(np.arange(0.5, 5, 1), minor=True)
ax.grid(which='minor')
ax.set_yticks(np.arange(0, 6, 1))
ax.set_yticks(np.arange(0, 6, 1))
ax.set_xlabel('Time [years]')
ax.set_ylabel('Digital growth [%]')
ax.set_title('Growth in Digital')
ax.legend(loc=2)
ax.grid()
```

```python
fig.clf()
fig, ax = plt.subplots()
ax.bar(['VR', 'Bots', 'Data-Centric', 'Digital'], [1, 2, 3, 4], color='C2')
ax.set_yticks(range(6))
ax.set_title('Count of Buzz Words in Presentation')
ax.set_ylim((0, 5))
ax.set_ylabel('Count')
ax.grid()
```

Subplots can also be useful 

```python
x = np.random.randint(low=1, high=11, size=50)
y = x + np.random.randint(1, 5, size=x.size)
data = np.column_stack((x, y))

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2,
                               figsize=(8, 4))

ax1.scatter(x=x, y=y, marker='o', c='r', edgecolor='b')
ax1.set_title('Scatter: $x$ versus $y$')
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')

ax2.hist(data, bins=np.arange(data.min(), data.max()), label=('x', 'y'))
ax2.legend(loc=(0.65, 0.8))
ax2.set_title('Frequencies of $x$ and $y$')
ax2.yaxis.tick_right()
fig.savefig('sub.png')
```

Sometimes is it better to save graphs as pdf when you have many graphs to create

```python
from matplotlib.backends.backend_pdf import PdfPages

with PdfPages('multipage_pdf.pdf') as pdf:
    plt.figure(figsize=(3, 3))
    plt.plot(range(7), [3, 1, 4, 1, 5, 9, 2])
    plt.title('Page One')
    pdf.savefig()  # saves the current figure into a pdf page
```
